My linear fit produced a slope of -.924*( log( Earthquake Frequency) / Earthquake magnitude).

As we can see in the plot below, as the magnitude approaches larger numbers (>5), the residual becomes larger. This could be improved by increasing the degree of polynomials we fit the data too, which increases model complexity and increases accuracy in our "training set"

```python
def fitpoly_degN(x,y,n):
    xl= x.size
    X = np.zeros((xl,n+1))
    for i in range(xl):
        for j in range(n+1):
            X[i,j]= (x[i]**j)
    X = np.matrix(X)
    AA = X.T@X
    BB = X.T@y
    A = np.linalg.solve(AA,BB.T)
    return np.asarray(A,dtype=float)

def evalpoly(coefs,point):
    deg = coefs.size
    Y = 0
    for i in range(deg):
        Y += coefs[i]*(point**i)
    return Y
```

```python
quakefit = fitpoly_degN(np.array(magsizebincenter),np.array(magfreq_lg10),1)
line = np.linspace(min(magsizebincenter),max(magsizebincenter),300)
linefit = evalpoly(quakefit,line)
plt.plot(magsizebincenter, magfreq_lg10, 'ro')
plt.plot(line, linefit, label = 'SLOPE =  '+ str(quakefit[1,0]))
plt.title("Line Fitting of Earthquake Data with Degree 1 Polynomial")
plt.xlabel('Earthquake Magnitude')
plt.ylabel('Earthquake Frequency (Log 10)')
plt.legend()
plt.show()
```
